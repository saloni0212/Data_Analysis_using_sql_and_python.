```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("orders.csv", na_values=['Not Available','unknown']) # anything inside na_values=[] will be considered as NaN
```

```python
#df.rename(columns={'Order Id':'order_id', 'City':'city'})

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')

df.head(5)
```

```python
# derive new columns discount, sale price and profit

df['discount'] = df['list_price']*df['discount_percent']*0.1
df['sale_price'] = df['list_price'] - df['cost_price']
df['profit'] = df['sale_price'] - df['cost_price']
df.head()
```

```python
# Convert order date from object data type to date time

df['order_date'] = pd.to_datetime(df['order_date'], format="%Y-%m-%d")
df.head()
```

```python
df.drop(columns=['list_price','cost_price','discount_percent'], inplace=True)
df.head()
```

```python
df.info()
```

```python
# Load the data into sql server using replace option

from sqlalchemy import create_engine

# Connection details
username = 'root'
password = 'root'
server = 'localhost'
port = '3306'
dbname = 'sakila'

# Create the connection URL
connection_url = f'mysql+pymysql://{username}:{password}@{server}:{port}/{dbname}'

# Create the engine
engine = create_engine(connection_url)

# Test the connection
try:
    connection = engine.connect()
    print("Connection successful!")
    connection.close()
except Exception as e:
    print(f"Connection failed: {e}")

# Load the data into sql server using append option
df.to_sql('df', con=engine , index=False, if_exists='append')
```
